```python
import outp_parser
import json
```

### Tally Parsing and Saving Example

```python
# Find file.
outp_path = "c100w_plateflux_production_09052025.outp"
# Make a tallies dictionary from output.
tally_dict = outp_parser.get_tallies(outp_path)
# Check to make sure the first key can be accessed.
tally_dict['Assembly MTR-F-001, Element #1']
```

```python
# Produce desired filepath to JSON file.
json_path = "c100w_plateflux_production_09052025.outp"
# Dump the dictionary to the JSON file.
with open(json_path, 'w') as f:
    json.dump(tally_dict, f)
```

### Tally Loading

```python
# Load the JSON file.
with open(json_path, 'r') as jsonfile:
    loaded_tallies = json.load(jsonfile)
# Check to see if the first key is present.
loaded_tallies['Assembly MTR-F-001, Element #1']
```

```python
# Load the tallies dictionary as DataFrames.
tally_dfs = outp_parser.conv_tallies(loaded_tallies)
# Check if the first key is a DataFrame now.
tally_dfs['Assembly MTR-F-001, Element #1']
```

### Handling

```python
# Get the keys from a dictionary.
keys = tally_dfs.keys()
# Check the keys.
keys
```

```python
# Cycle through all keys in a dictionary.
tallies_data = []
for key in keys:
    this_tally = tally_dfs[key]
    tallies_data.append(this_tally)
# Check to make sure all the tallies were pulled.
len(tallies_data)
```

```python
# Convert a DataFrame to a numpy array.
tally_np = tally_dfs['Assembly MTR-F-001, Element #1'].to_numpy()
# Check it.
tally_np
```

```python
# Pull the energy 'column' from numpy.
# Note that '.T' is a transpose operation.
energies = tally_np.T[0]
```

```python
# Pull the energy 'column' from the DataFrame as a numpy array.
energies = tally_dfs['Assembly MTR-F-001, Element #1']['energy'].to_numpy()
# Check it.
energies
```
